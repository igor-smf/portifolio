```python
# Fonte - https://datacatalog.worldbank.org/search/dataset/0037712/World-Development-Indicators
path = '/content/drive/MyDrive/WDIData.csv'
df = pd.read_csv(path)
df
```

```python
# Selecionar apenas as linhas em que o país é o Brasil
df1 = df[df["Country Name"] == 'Brazil']
df1
```

```python
# Criar uma lista de strings com anos de 1985 até 2022
anos_desejados = [str(ano) for ano in range(1985, 2023)]

# Exibir a lista
print(anos_desejados)
```

```python
# Selecionar apenas as colunas "Country Name", "Indicator Name" e as colunas especificadas em anos_desejados
df1 = df1[['Country Name', 'Indicator Name'] + anos_desejados]
```

**Sugestõesde indicadores para análise:** 'GDP (current US$)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)',
'Inflation, consumer prices (annual %)', 'Life expectancy at birth, total (years)', 'CO2 emissions (metric tons per capita)',
'Foreign direct investment, net outflows (% of GDP)', 'Gini index (World Bank estimate)'

```python
indicator_name = 'GDP (current US$)'
```

```python
# Selecionando apenas as linhas de df1 onde o "Indicator Name" é igual ao valor fornecido em indicator_name
df1_indicator = df1[df1["Indicator Name"] == indicator_name ]
df1_indicator
```

```python
# Realizar a transposição do DataFrame df1_indicator. A transposição é uma operação que troca as linhas pelas colunas e vice-versa
df1_indicator = df1_indicator.T
```

```python
# Excluir valores nulos
df1_indicator = df1_indicator.dropna()
```

```python
# Converter os valores para números
df1_indicator = df1_indicator.iloc[2:].apply(pd.to_numeric)

# Obter os anos e os valores do GDP
anos = df1_indicator.index.astype(int)
values = df1_indicator.iloc[:, 0].values
```

```python
from itertools import cycle
import locale

# Configurar a formatação de milhares
locale.setlocale(locale.LC_NUMERIC, 'en_US.UTF-8')  # Usando o formato en_US como exemplo

# Plotar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(anos, values, marker='o', label='GDP')
plt.title(f'{indicator_name} do Brasil ao longo do tempo')
plt.xlabel('Ano')
plt.ylabel(indicator_name)

# Definir um ciclo de cores para os presidentes
cores_presidentes = cycle(['b', 'g', 'r', 'c', 'm', 'y', 'k'])

# Adicionar linhas verticais para indicar os anos de início de cada presidente com cores diferentes
presidentes = {
    'Collor': 1990,
    'Itamar': 1992,
    'FHC': 1995,
    'Lula': 2003,
    'Dilma': 2011,
    'Temer': 2016,
    'Bolsonaro': 2019
}

for presidente, ano_inicio in presidentes.items():
    cor = next(cores_presidentes)
    plt.axvline(x=ano_inicio, color=cor, linestyle='--', label=presidente, alpha=0.7)

# Adicionar linhas horizontais para indicar máximo, mínimo e mediana
max_value = max(values)
min_value = min(values)
median_value = np.median(values)
ano_max = anos[values.argmax()]
ano_min = anos[values.argmin()]

max_value = '{:,.2f}'.format(max_value).replace(',', ' ').replace('.', ',')
min_value = '{:,.2f}'.format(min_value).replace(',', ' ').replace('.', ',')
median_value = '{:,.2f}'.format(median_value).replace(',', ' ').replace('.', ',')

# Adicionar informações na legenda
legenda = (
    f'Máximo: {max_value} (Ano: {ano_max})\n'
    f'Mínimo: {min_value} (Ano: {ano_min})\n'
    f'Mediana: {median_value}'
)
plt.legend(title=legenda, loc='upper right', bbox_to_anchor=(1.5, 1.25))

plt.grid(True)
plt.show()
```
